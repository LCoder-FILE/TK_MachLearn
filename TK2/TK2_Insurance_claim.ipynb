```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

data = pd.read_csv("insurance.csv")
y = data['charges'].values
X = data[['age', 'bmi', 'children']].values

X_y = data.drop(columns=['sex','smoker','region'])

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = X_y.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
new_data = X_test[-1:]
new_data
```

```python
predict = model.predict(new_data)
print("Predicted : ", predict)
print("Actual : ", y_test[-1:])
```

```python
y_pred = model.predict(X_test)
cost_function_pred = np.sum(((y_pred - y_test) ** 2) / (2 * len(y_test)))
cost_function_pred
```

```python
y_pred_train = model.predict(X_train)
cost_function_train = np.sum(((y_pred_train - y_train) ** 2) / (2 * len(y_train)))
cost_function_train
```

### Conclusion : 

From the model score and low valued correlation matrix (close to 0), we can conclude that the variables in the data not having strong linear relationship.
